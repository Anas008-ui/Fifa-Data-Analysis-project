```python
#To plot the histogram figures for the distribution of ages
plt.figure(figsize=(10, 6))
sns.histplot(fifa['age'], bins=20, kde=True)
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#To print the top club by players
valuable_clubs = fifa.groupby('club')['value_eur'].count().sort_values(ascending=False).head(10)
valuable_clubs.plot(kind='bar', figsize=(8, 6), title='Top Clubs by Number of Players')
plt.ylabel('Number of Players')
plt.show()
```

```python
# Average player rating by nationality (Top 10)
avg_rating_by_country = (
    fifa.groupby('nationality')['overall']
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .round(2)
)

plt.figure(figsize=(10, 6))
sns.barplot(
    x=avg_rating_by_country.values,
    y=avg_rating_by_country.index,
    palette="magma",
    hue=avg_rating_by_country.index, # Explicitly set hue to the y-axis values
    legend=False # Turn off the legend since hue is the y-axis
)

# Add text labels for clarity
for index, value in enumerate(avg_rating_by_country.values):
    plt.text(value + 0.1, index, str(value), va='center', fontsize=9, color='black')

plt.title("Top 10 Nationalities by Average Player Rating", fontsize=14, fontweight='bold')
plt.xlabel("Average Overall Rating")
plt.ylabel("Nationality")
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Top 10 clubs by number of players
valuable_clubs = (
    fifa.groupby('club')['value_eur']
    .count()
    .sort_values(ascending=False)
    .head(10)
)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(
    x=valuable_clubs.values,
    y=valuable_clubs.index,
    palette="coolwarm",
    hue=valuable_clubs.index, # Explicitly set hue to the y-axis values
    legend=False # Turn off the legend since hue is the y-axis
)
plt.title('Top 10 Clubs by Number of Players', fontsize=14, fontweight='bold')
plt.xlabel('Number of Players')
plt.ylabel('Club')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#To get the Wage and overall rating of Players
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=fifa,
    x='overall',
    y='wage_eur',
    hue='international_reputation',
    palette='cool',
    alpha=0.6
)
plt.title("Wage vs Overall Rating of Players", fontsize=14, fontweight='bold')
plt.xlabel("Overall Rating")
plt.ylabel("Wage (in Euros)")
plt.legend(title="International Reputation")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
 #To show  attacking players of top clubs only,i.e,(FC Barcelona''Real Madrid','Manchester City','Juventus','Liverpool','Paris Saint-Germain','Bayern München')
 import pandas as pd

# Load the CSV file
file_path = "players_20 (1).csv"  # Update this if needed
df = pd.read_csv(file_path)

# Define attacking positions
attacking_positions = ['ST', 'CF', 'LW', 'RW', 'LF', 'RF', 'CAM']

# Define your list of good clubs
good_clubs = [
    'FC Barcelona',
    'Real Madrid',
    'Manchester City',
    'Juventus',
    'Liverpool',
    'Paris Saint-Germain',
    'Bayern München'
]

# Function to check if a player has any attacking position
def is_attacking(player_positions):
    if pd.isnull(player_positions):
        return False
    for pos in player_positions.split(', '):
        if pos in attacking_positions:
            return True
    return False

# Filter for attacking positions
attacking_players = df[
    df['player_positions'].apply(is_attacking)
]

# Further filter for good clubs
attacking_players_from_good_clubs = attacking_players[
    attacking_players['club'].isin(good_clubs)
]

# Show player name, club, position
print(attacking_players_from_good_clubs[
    ['short_name', 'club', 'player_positions']
])
```
